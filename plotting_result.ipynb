```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
delta = np.array([1, 2, 3, 4])
wdgrl_parametric = np.array([0.099173554, 0.223140496, 0.553719008264462, 0.809917355])
wdgrl_oc = np.array([0.074380165, 0.213675214, 0.495867768595041, 0.396694215])

# Plot the data
plt.plot(delta, wdgrl_parametric, marker='o', linestyle='-', label="WDGRL-parametric", color='tab:blue')
plt.plot(delta, wdgrl_oc, marker='o', linestyle='-', label="WDGRL-oc", color='tab:orange')

plt.xlabel("Delta")
plt.ylabel("TPR")
plt.xticks(delta)
plt.yticks([0.0, 0.2, 0.4, 0.6, 0.8])  

# Legend
plt.legend()

# Show the plot
plt.show()
```
