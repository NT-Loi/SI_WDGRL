```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
delta = np.array([1, 2, 3, 4])
wdgrl_parametric = np.array([0.090909091, 0.241666666666666, 0.504132231404958, 0.851239669421487])
wdgrl_oc = np.array([0.0909090909090909, 0.190082644628099, 0.272727272727272, 0.330578512396694])

# Plot the data
plt.plot(delta, wdgrl_parametric, marker='o', linestyle='-', label="WDGRL-parametric", color='tab:blue')
plt.plot(delta, wdgrl_oc, marker='o', linestyle='-', label="WDGRL-oc", color='tab:orange')

plt.xlabel("Delta")
plt.ylabel("TPR")
plt.xticks(delta)
plt.yticks([0.0, 0.2, 0.4, 0.6, 0.8])  

# Legend
plt.legend()

# Show the plot
plt.show()
```
